### d)	Hangi genler «lymphoid_neoplasm»  hastalığında etkin olmuş. Gen histogramını çıkaralım. 


```python
filtered_df2 = df[["Primary histology","Gene name"]].copy()
filtered_df2
```

```python
lymp = pd.DataFrame(filtered_df2.loc[filtered_df2["Primary histology"] == "lymphoid_neoplasm", "Gene name"])
lymp
```

```python
pd.options.display.max_rows = 707 # 707 unique gene var
gene_counts = lymp["Gene name"]
Lymp_genes = pd.DataFrame(gene_counts.value_counts())
Lymp_genes # lymphoid_neoplasm için gene histogramı çıkarılır
```
